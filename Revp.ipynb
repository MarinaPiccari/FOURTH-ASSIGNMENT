```python
def read_fasta(file):
    NL = []
    L = file.read().split('>')[1:]
    for i in range(len(L)):
        new = L[i].split('\n', 1)
        seq = new[1].replace('\n', '')
        NL.append(seq)
    return (NL)

def complementary_strand(string):
    revs = string[::-1]
    revc = []
    for x in range(len(revs)):
        if revs[x] == 'T':
            revc.append('A')
        elif revs[x] == 'A':
            revc.append('T')
        elif revs[x] == 'C':
            revc.append('G')
        else:
            revc.append('C')
    return ''.join(revc)
    
    
with open ('rosalind_revp.fasta', 'r') as f:
    seq = read_fasta(f)[0]
    for i in range(len(seq)):
        for j in range(i+1, len(seq)+1):
            L = []
            temp = str(seq[i:j])
            if 4 <= len(temp) <= 12:
                if temp[::-1] == complementary_strand(temp[::-1]):
                    L.extend([i+1, j-i])
                    print (*L)
```
