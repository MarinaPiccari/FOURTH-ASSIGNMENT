```python
def read_fasta(file):
    NL = []
    L = file.read().split('>')[1:]
    for i in range(len(L)):
        new = L[i].split('\n', 1)
        seq = new[1].replace('\n', '')
        NL.append(seq)
    return (NL)

with open ('rosalind_cons.fasta', 'r') as f:
    L = read_fasta(f)
    l = len(L[0])
    A = []
    C = []
    G = []
    T = []
    d = []
    
    for i in range(l):
        a = c = g = t = 0
        for j in L:
            if j[i] == 'A':
                a += 1
            elif j[i] == 'C':
                c += 1
            elif j[i] == 'G':
                g += 1
            elif j[i] == 'T':
                t += 1
                
        A.append(a)
        C.append(c)
        G.append(g)
        T.append(t)
        
    d.extend([A, C, G, T])
    cons = []
    for i in range(l):
        m = []
        for lst in d:
            m.append(lst[i])
        M = max(m)
        if m.index(M) == 0:
            cons.append('A')
        elif m.index(M) == 1:
            cons.append('C')
        elif m.index(M) == 2:
            cons.append('G')
        elif m.index(M) == 3:
            cons.append('T')
            
    
    print (''.join(cons))
    print ('A:', *A)
    print ('C:', *C)
    print ('G:', *G)
    print ('T:', *T)
```
