## Longest increasing subsequence
##### Marina Piccari
#### 1904900

```python
with open ('rosalind_lgis.txt', 'r') as f:
    n = int(f.readline())
    L = list(map(int, f.readline().split()))
    
    incr = [(0,[])]*(n+1)
    decr = [(0,[])]*(n+1)
    
    for i in L:
        x,y = max(incr[:i], key = lambda j : j[0])
        incr[i] = (x+1,y+[i])
    sbincr = max(incr)[1]
    
    for i in L:
        x,y = max(decr[i:], key = lambda j : j[0])
        decr[i] = (x+1, y+[i])
    sbdecr = max(decr)[1]
    
    print (*sbincr)
    print (*sbdecr)
```
